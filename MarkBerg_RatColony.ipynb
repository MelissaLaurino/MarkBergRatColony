```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df=rats

# Setting the positions and width for the bars
pos = list(range(len(df['Weight_g']))) 
width = 0.45 
    
# Plotting the bars
fig, ax = plt.subplots(figsize=(20,10))

# Create a bar with pre_score data,
# in position pos,
plt.bar(pos, 
        #using df['pre_score'] data,
        df['Weight_g'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#EE3224', 
        # with label the first value in first_name
        label=df['RATID'][0]) 

# Create a bar with mid_score data,
# in position pos + some width buffer,
plt.bar([p + width for p in pos], 
        #using df['mid_score'] data,
        df['Weight_g'],
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#F78F1E', 
        # with label the second value in first_name
        label=df['RATID'][1]) 

# Create a bar with post_score data,
# in position pos + some width buffer,
plt.bar([p + width*2 for p in pos], 
        #using df['post_score'] data,
        df['Weight_g'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#FFC222', 
        # with label the third value in first_name
        label=df['RATID'][2]) 

# Set the y axis label
ax.set_ylabel('Weights (g)')

# Set the chart's title
ax.set_title('Mark Berg Rat Colony')

# Set the position of the x ticks
ax.set_xticks([p + 1.5 * width for p in pos])

# Set the labels for the x ticks
ax.set_xticklabels(df['RATID'])

# Setting the x-axis and y-axis limits
plt.xlim(min(pos)-width, max(pos)+width*10)
#plt.ylim([0, max(df['Weight_g'] + df['Weight_g'] + df['Weight_g'])] )

# Adding the legend and showing the plot
plt.legend(['', '', ''], loc='upper left')
plt.grid()
plt.show()
```

```python
import pylab
import math #Linear Regression
def  getX(y, a, b):
    return ((y-b)/a)
     
def getY(x, a, b):
    return ((a*x) + b)
    
def readFile(filename):
    data = (open(filename, "r")).readlines()
    new = []
    for i in data:
        new.append(i.split())
        
    brain = []
    body = []
    for i in new:
        brain.append(i[1])
        body.append(i[0])
    
    brain = brain[1:]
    body =  body[1:]
    
    for i in range (len(brain)):
        brain[i] = float(brain[i])
        body[i] = float(body[i])
    
    print(brain,body)
    return (brain,body)
```

```python
(y,x) = readFile("Rat1.txt")

x = pylab.array(x)
y = pylab.array(y)

Min = min(x)
Max = max(x)

x1 = np.linspace(Min, Max, 100000)
m,c = pylab.polyfit(x,y,1)
#y1 = pylab.polyval(model,x)
#y = a*x + b
y1 = (m*x1) + c

print("y = ", m,"x + ",c)

pylab.plot(x,y, "bo", label = 'measured points')
pylab.plot(x1,y1, "r-", label = 'Linear fit')
pylab.legend(loc = "upper left")
pylab.xlabel("Body mass (kg)")
pylab.ylabel("brain mass (g)")
pylab.grid()
pylab.show()
```
