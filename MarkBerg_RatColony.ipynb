```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df=rats

# Setting the positions and width for the bars
pos = list(range(len(df['Weight_g']))) 
width = 0.45 
    
# Plotting the bars
fig, ax = plt.subplots(figsize=(20,10))

# Create a bar with pre_score data,
# in position pos,
plt.bar(pos, 
        #using df['pre_score'] data,
        df['Weight_g'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#EE3224', 
        # with label the first value in first_name
        label=df['RATID'][0]) 

# Create a bar with mid_score data,
# in position pos + some width buffer,
plt.bar([p + width for p in pos], 
        #using df['mid_score'] data,
        df['Weight_g'],
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#F78F1E', 
        # with label the second value in first_name
        label=df['RATID'][1]) 

# Create a bar with post_score data,
# in position pos + some width buffer,
plt.bar([p + width*2 for p in pos], 
        #using df['post_score'] data,
        df['Weight_g'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#FFC222', 
        # with label the third value in first_name
        label=df['RATID'][2]) 

# Set the y axis label
ax.set_ylabel('Weights (g)')

# Set the chart's title
ax.set_title('Mark Berg Rat Colony')

# Set the position of the x ticks
ax.set_xticks([p + 1.5 * width for p in pos])

# Set the labels for the x ticks
ax.set_xticklabels(df['RATID'])

# Setting the x-axis and y-axis limits
plt.xlim(min(pos)-width, max(pos)+width*10)
#plt.ylim([0, max(df['Weight_g'] + df['Weight_g'] + df['Weight_g'])] )

# Adding the legend and showing the plot
plt.legend(['', '', ''], loc='upper left')
plt.grid()
plt.show()
```
